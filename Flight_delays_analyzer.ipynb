```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/flight_delays_2020.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data =data[data['CANCELLED'] == 0]
```

```python
Data = data.drop(columns=['TAIL_NUM','CANCELLATION_CODE'])
```

```python
data['DEP_HOUR'] = data['CRS_DEP_TIME']
```

```python
data["DEP_HOUR"] = data["DEP_HOUR"].astype(int)
```

```python
data["Shift"] = data["DEP_HOUR"].apply(lambda x: 'Morning' if x < 12 else 'Afternoon')
```

```python
cleaned_data = data[['FL_DATE','MONTH','DAY_OF_WEEK','ORIGIN','DEST','MKT_UNIQUE_CARRIER',
                     'DEP_DELAY','ARR_DELAY','DEP_HOUR','Shift']]
```

```python
cleaned_data.head()
```

```python
print("\nMissing values after cleaning:")
print(cleaned_data.isnull().sum())
```

```python
cleaned_data = cleaned_data.dropna(subset=['ARR_DELAY'])
```

```python
print("Missing values after removing ARR_DELAY NaNs:")
cleaned_data.isnull().sum()
```

```python
cleaned_data.head()
```

# Most Delayed Airports (Bar Chart)

```python
# Group by origin airport and calculate mean departure delay
airport_delays = cleaned_data.groupby('ORIGIN')['DEP_DELAY'].mean().sort_values(ascending=False)
```

```python
airport_delays
```

```python
# Plot top 15 most delayed airports
plt.figure(figsize=(12,6))
airport_delays.head(15).plot(kind='bar', color='salmon')
plt.title("Top 15 Most Delayed Origin Airports")
plt.xlabel("Airport")
plt.ylabel("Average Departure Delay (minutes)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
```

# Monthly Delay Trends (Time-Series Plot)

```python
# Average monthly delay
monthly_delays = cleaned_data.groupby('MONTH')['DEP_DELAY'].mean()
monthly_delays
```

```python
plt.figure(figsize=(10,5))
monthly_delays.plot(marker='o', linestyle='-')
plt.title("Average Departure Delay by Month")
plt.xlabel("Month")
plt.ylabel("Average Delay in Minutes")
plt.grid(True)
plt.xticks(range(1,11))
plt.tight_layout()
```

# Morning vs Evening Delay Comparision (Boxplot)

```python
plt.figure(figsize=(8,6))

sns.boxplot(x='Shift', y='DEP_DELAY', data = cleaned_data)
plt.title("Departure Delays: Morning vs Afternoon")
plt.xlabel("Shift")
plt.ylabel("Delay in minutes")
plt.tight_layout()
```

# Airport vs Airline Delay Heatmap

```python
plt.figure(figsize=(14,8))

#Pivot table
heatmap_data = cleaned_data.pivot_table(
    values = 'DEP_DELAY',
    index = 'ORIGIN',
    columns = 'MKT_UNIQUE_CARRIER',
    aggfunc = 'mean'
)

#plot heatmap
sns.heatmap(heatmap_data, cmap='YlOrRd', linewidths=0.5)
plt.title("Average Departure Delay: Airport vs Airline")
plt.xlabel("Airline")
plt.ylabel("Origin Airport")
plt.tight_layout()
```

# Morning vs Evening Delay % Difference

```python
morning_avg = cleaned_data[cleaned_data['Shift'] == 'Morning']['DEP_DELAY'].mean()
morning_avg
```

```python
evening_avg = cleaned_data[cleaned_data['Shift'] == 'Afternoon'] ['DEP_DELAY'].mean()
evening_avg
```

```python
difference_minutes = evening_avg - morning_avg
percent_increase = (difference_minutes / abs(morning_avg)) * 100
percent_increase
```

```python
print(f"Average Morning Delay: {morning_avg:.2f} minutes")
print(f"Average Evening Delay: {evening_avg:.2f} minutes")
print(f"Evening flights are delayed by {difference_minutes:.2f}% more minutes than morning flights.")
print(f"That is a {percent_increase:.2f}% increase relative to morning flights.")
```
